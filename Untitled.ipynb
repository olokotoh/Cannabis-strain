```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
```

```python
df = pd.read_csv('Cannabis StrainsFeatures - cannabis.csv')
df.head()
```

```python
df.info()
print(df.shape)
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(6,6))
g = sns.countplot(x="Type" ,data=df, 
                  palette='hls')
plt.show()
```

```python
rating_by_type = df.groupby('Type')['Rating'].count()
rating_by_type
```

```python
plt.figure(figsize=(15,10))

plt.suptitle('Rating Distributions by Cannabis Species', fontsize=22)

#Let's look the Rating distribuition by Type.
g = plt.subplot(231)
g = sns.distplot(df[(df.Type == 'hybrid') & 
                               (df.Rating > 0)]["Rating"],
                 color='y')
g.set_xlabel("Rating", fontsize=15)
g.set_ylabel("Probability", fontsize=15)
g.set_title("Hybrids", fontsize=20)

g1 = plt.subplot(232)
g1 = sns.distplot(df[(df.Type == 'sativa') & 
                               (df.Rating > 0)]["Rating"],
                  color='g')
g1.set_xlabel("Rating", fontsize=15)
g1.set_ylabel("Probability", fontsize=15)
g1.set_title("Sativas",
             fontsize=20)

g2 = plt.subplot(233)
g2 = sns.distplot(df[(df.Type == 'indica') & 
                               (df.Rating > 0)]["Rating"], color='r')
g2.set_xlabel("Rating", fontsize=15)
g2.set_ylabel("Probability", fontsize=15)
g2.set_title("Indicas", 
             fontsize=20)

g3 = plt.subplot(212)

my_pal = {"hybrid": "y", "sativa": "g", "indica":"r"}

#I will now explore the Rating distribuition by Type
g3 = sns.boxplot(x="Type",y="Rating",  palette=my_pal,
                data=df[df["Rating"] > 2])
g3.set_title("Boxplot of Distribuition by Species Type", 
            fontsize=20)
g3.set_xlabel("Species", fontsize=17)
g3.set_ylabel("Rating", fontsize=17)

plt.subplots_adjust(wspace = 0.3, hspace = 0.4,top = 0.85)

plt.show()
```
